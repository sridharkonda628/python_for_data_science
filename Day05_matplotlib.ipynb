```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

xpoints=np.array([0,6])
ypoints=np.array([0,10])

plt.figure(figsize=(1,1))
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,6])
ypoints=np.array([3,25])

plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
xpoints=np.array([0,6,7,8,9,25])

#use of marker at points
plt.plot(xpoints,marker='o')
plt.show()
```

```python
xpoints=np.array([0,6,7,88,45,5,55,68,2,4,2,65,43,24,38,52,49,25])


plt.plot(xpoints,marker='*')
plt.show()
```

```python
xpoints=np.array([0,6,7,88,45,5,55,68,2,4,2,65,43,24,38,52,49,25])

# use of line style

plt.plot(xpoints,marker='*',linestyle=':')
plt.show()
```

```python
plt.figure(figsize=(3,3))
xpoints=np.array([0,6,7,88,45,5,55,68,2,4,2,65,43,24,38,52,49,25])

plt.title('betting earnings report')
plt.xlabel('money spent')
plt.ylabel('profits earned')
#ms=marker size, mfc=marker face color, mec=marker edge color
plt.plot(xpoints,marker='*', mfc='pink', mec='red')
#to add grid
plt.grid(color='green',ls=':',linewidth=0.7)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(xpoints)
plt.title('top')
plt.show()

plt.subplot(1,2,1)
plt.plot(ypoints)
plt.title('bottom')
plt.supertitle('naruto')

plt.show()
```

```python
x1=np.array([5,1,2,3,4,3,8,6,1,4,5,6,])
y1=np.array([6,1,5,3,6,4,8,6,3,5,3,4,])
plt.bar(x1,y1,color='green',width=0.8)
plt.title('sports report')
plt.xlabel('year')
plt.ylabel('no.of wins')
plt.show()
```

```python
x1=np.array([5,1,2,3,4,3,8,6,1,4,5,6,])
y1=np.array([6,1,5,3,6,4,8,6,3,5,3,4,])
#
#horizontal representation of bar graphs
plt.barh(x1,y1)
plt.title('sports report',fontsize=20)
plt.xlabel('year')
plt.ylabel('no.of wins')
plt.show()
```

```python
plt.scatter(x1,y1)

plt.colorbar()
plt.show()
```
